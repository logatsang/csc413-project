```python
import pandas as pd
```

```python
df = pd.read_csv("data/etymology.csv")
df.head()
```

```python
eng_df = df[(df['lang'] == 'English') & (df['reltype'] == 'borrowed_from')]
eng_df = eng_df[['term', 'related_lang']].sort_values('term')
eng_df.drop_duplicates(subset=['term'], keep=False)
eng_df.head()
```

```python
eng_df.to_csv('data/etymology_proc.csv', index=False)
```
